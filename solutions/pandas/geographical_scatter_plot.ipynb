## Importing pandas and matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reading my file

## In this dataset, we have price instead of population

```python
df = pd.read_csv('optimised_food_price.csv')
df.head()
```

```python
df.columns
```

```python
df['commodity'].unique()
```

## Filtering the markets having tomato price greater than 80 rupees

```python
df_tomato_price_gt_80 = df.loc[(df['commodity']=='Tomatoes')&(df['price']>80)]
```

```python
df_tomato_price_gt_80
```

## Filtering market,latitude and longitude features alone

```python
df_market_geo = df_tomato_price_gt_80[['market', 'latitude', 'longitude']].drop_duplicates()
```

```python
df_market_geo
```

## Creating geographical plot using matplotlib

```python
# Creating a scatter plot
fig = plt.figure(dpi = 120, figsize=(12,6))
ax = fig.add_subplot(1,1,1)
ax.scatter(df_market_geo['longitude'], df_market_geo['latitude'], s=50, c='blue', alpha=0.5, edgecolors='k')

# Add labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Geographical Scatter Plot of markets')

#Annotate data points with city names.
for index, market in df_market_geo.iterrows():
    plt.annotate(market['market'], (market['longitude'], market['latitude']), textcoords='offset points', xytext=(0, 5), ha='center')
```
