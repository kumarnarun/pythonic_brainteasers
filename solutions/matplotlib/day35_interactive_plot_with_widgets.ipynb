## Challenge: Create an interactive Matplotlib plot with widgets (e.g., sliders, checkboxes) to allow users to customize the appearance of the plot, such as adjusting the time range or filtering data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import ipywidgets as widgets
```

## Reading my file

```python
df = pd.read_csv('optimised_food_price.csv')
df.head()
```

## Filtering the tomatoes from the whole commodity of Indian foods

```python
df_tomato = df.loc[(df['commodity']=='Tomatoes')]
```

```python
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(df_tomato['longitude'], df_tomato['latitude'], s=50, c='green', alpha=0.7, edgecolors='k')

# Add labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Geographical Scatter Plot of markets')

plt.subplots_adjust(bottom=0.25)    
ax_slider = plt.axes([0.1,0.1,0.8,0.05],facecolor = 'teal')
slider = Slider(ax_slider,"Price", valmin=1,valmax=110,valinit=50,valstep=1)
slider.on_changed(update_scatter)

plt.show()
```
