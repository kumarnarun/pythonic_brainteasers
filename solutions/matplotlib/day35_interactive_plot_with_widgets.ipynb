## Filtering the markets having tomato price greater than 80 rupees

```python
df_tomato = df.loc[(df['commodity']=='Tomatoes')]
```

```python
df_tomato
```

```python
fig, ax = plt.subplots(figsize = (12,8))
ax.scatter(df_tomato['longitude'], df_tomato['latitude'], s=50, c='green', alpha=0.7, edgecolors='k')

# Add labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Geographical Scatter Plot of markets')

plt.subplots_adjust(bottom=0.25)


def update_scatter(val):
    ax.clear()
    df_price = df_tomato.loc[df_tomato['price']==val]
    ax.scatter(df_price['longitude'], df_price['latitude'], s=50, c='green', alpha=0.7, edgecolors='k')
    ax.set_title(f'Places where tomato price is {val}')
    plt.draw()
    
ax_slider = plt.axes([0.1,0.1,0.8,0.05],facecolor = 'teal')
slider = Slider(ax_slider,"Price", valmin=1,valmax=110,valinit=50,valstep=1)
slider.on_changed(update_scatter)

plt.show()
```
